```python
# Enable interactive plot
%matplotlib notebook
%matplotlib notebook

import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import numpy as np
import pandas as pd

plt.style.use("ggplot")
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(10, 5))

line, = axes.plot([])

plt.xlabel("Batch")
plt.ylabel("NÃºmero de Palavras (a cada 5 segundos)")

#red_patch = mpatches.Patch(color='red', label='The red data')

def animate(frame_num):
    dataP = pd.read_csv('../palavrasComP.csv')
    dataS = pd.read_csv('../palavrasComS.csv')
    dataR = pd.read_csv('../palavrasComR.csv')
    data6 = pd.read_csv('../palavrasCom6.csv')
    data8 = pd.read_csv('../palavrasCom8.csv')
    data11 = pd.read_csv('../palavrasCom11.csv')
    
    p, = axes.plot(dataP['batch'], dataP['palavrasComP'], color="green", label="Palavras com P")
    r, = axes.plot(dataR['batch'], dataR['palavrasComR'], color="gray", label="Palavras com R")
    s, = axes.plot(dataS['batch'], dataS['palavrasComS'], color="blue", label="Palavras com S")
    seis, = axes.plot(data6['batch'], data6['palavrasCom6'], color="red", label="Palavras com 6 char.")
    oito, = axes.plot(data8['batch'], data8['palavrasCom8'], color="yellow", label="Palavras com 8 char.")
    onze, = axes.plot(data11['batch'], data11['palavrasCom11'], color="orange", label="Palavras com 11 char.")
    axes.legend(handles=[p, r, s, seis, oito, onze])
    
    return line

anim = FuncAnimation(fig, animate, frames=100, interval=20)
```
